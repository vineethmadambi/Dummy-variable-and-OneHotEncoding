```python
#here I have a file of prices of different cars based on their mileage. Here I first plot this data on a scatter plot chart to
# see if i can use LinearRegression model.
#1) predict price of a mercedez benz that is 4 year old with mileage 45000
#2) predict price of a BMW X5 is 7 year old with mileage 86000
#3)find the score of my model.
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
df=pd.read_csv('carprices.csv')
df
```

```python
plt.scatter(df['Mileage'],df['Sell Price($)'])
```

```python
dummies=pd.get_dummies(df['Car Model'])
dummies
```

```python
merged=pd.concat([df,dummies],axis='columns')
merged
```

```python
final=merged.drop(['Car Model','Audi A5'],axis='columns')
final
```

```python
X=final.drop(['Sell Price($)'],axis='columns')
X
```

```python
y=final['Sell Price($)']
y
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.predict([[45000,4,0,1]])
```

```python
model.predict([[86000,7,1,0]])
```

```python
model.score(X,y)
```
